```python
import numpy as np
import pandas as pd
df = pd.read_csv("C:/Users/Aya Benrahla/Downloads/cars.csv")
# Show first rows
df.head()
# Dataset information
df.info()
# Basic statistics
df.describe()
# First numeric column
numeric_col = df.select_dtypes(include=[np.number]).columns[0]
numeric_col
arr = df[numeric_col].to_numpy()

print("First 10 values of NumPy array:")
print(arr[:10])

# Example NumPy operation
mean_value = np.mean(arr)
print("Mean value =", mean_value)
# Normalization
df['normalized'] = (df[numeric_col] - df[numeric_col].mean()) / df[numeric_col].std()

# Log transformation
df['log_value'] = np.log(df[numeric_col] + 1)

df.head()
import matplotlib.pyplot as plt

# Histogram
plt.figure(figsize=(6,4))
plt.hist(df[numeric_col], bins=20, color='skyblue')
plt.title("Histogram of " + numeric_col)
plt.xlabel(numeric_col)
plt.ylabel("Frequency")
plt.show()

# Line Plot (first 100 rows)
plt.figure(figsize=(6,4))
plt.plot(df[numeric_col][:100])
plt.title("Line Plot")
plt.xlabel("Index")
plt.ylabel(numeric_col)
plt.show()
# Second numeric column
other_numeric = df.select_dtypes(include=[np.number]).columns[1]

plt.figure(figsize=(6,4))
plt.scatter(df[numeric_col], df[other_numeric], color='green')
plt.title("Scatter Plot")
plt.xlabel(numeric_col)
plt.ylabel(other_numeric)
plt.show()
import plotly.express as px

# Histogram
fig = px.histogram(df, x=numeric_col, nbins=20, title="Plotly Histogram")
fig.show()

# Line plot
fig = px.line(df[numeric_col][:100], title="Plotly Line Plot")
fig.show()

# Scatter plot
fig = px.scatter(df, x=numeric_col, y=other_numeric, title="Plotly Scatter Plot")
fig.show()
# Example: load another file (change name as needed)
# df2 = pd.read_csv("C:/Users/Aya Benrahla/Downloads/another_dataset.csv")

# df2.head()
# px.scatter(df2, x=df2.columns[0], y=df2.columns[1], title="Other Dataset Example").show()
```
